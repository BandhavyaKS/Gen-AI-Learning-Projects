## **STEMMING**

## **(i) PorterStemmer**

```python
from nltk.stem import PorterStemmer
```

create an object for Porter stemmer(class)

```python
stemmer=PorterStemmer()
```

```python
words = [
    "Running",
    "jumps",
    "Easily",
    "fairly",
    "Leaves",
    "wolves",
    "children",
    "studies",
    "happier",
    "flying",
    "connected",
    "connections",
    "relational",
    "activation",
    "demonstration",
    "nationality",
    "computer",
    "caring",
    "better",
    "studying","a"
]
for word in words:
    print(f'{word}---> {stemmer.stem(word)}')
```

it does not convert into root form , it works based on rules
- it converts to lower case ()
- it removes suffic based on rules
- supports only english language

## **(ii) Snowball stemmer**

```python
# class SnowballStemmer(
#     language: str,
#     ignore_stopwords: bool = False
# )
# Snowball Stemmer

# The following languages are supported: Arabic, Danish, Dutch, English, Finnish, French, German, Hungarian, Italian, Norwegian, Portuguese, Romanian, Russian, Spanish and Swedish.

# The algorithm for English is documented here:

#     Porter, M. "An algorithm for suffix stripping."
#     Program 14.3 (1980): 130-137.
```

- supports multiple language
- snowball it uses ignores stoppper( ignores stoppers)

```python
from nltk.stem import SnowballStemmer
snowball_stemmer= SnowballStemmer( 'english',ignore_stopwords=True)
```

```python
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

stop_words=stopwords.words("english")
# stop_words

snowball_stemmer=SnowballStemmer("english",ignore_stopwords=True)
snowball_stemmer.stopwords=stop_words

snowball_stemmer.stopwords
```
