```python
from nltk import word_tokenize, sent_tokenize
```

- word_tokenize :-- using space as splitter
- sent_tokenize:--- uses "." as splitter

```python
word_tokenize("GenAI: The Natural Language Toolkit (NLTK) is an open source Python library for Natural Language Processing. A free online book is available. (If you use the library for academic research, please cite the book.)")
```

```python
sent_tokens=sent_tokenize("GenAI: The Natural Language. Toolkit (NLTK) is an open source Python library for Natural Language Processing. A free online book is available. (If you use the library for academic research, please cite the book.)")
```

```python
print(sent_tokens)
```

```python
res="GenAI: The Natural Langu           age. Toolkit (NLTK) is an open source Python library for Natural Language Processing. A free online book is available. (If you use the library for academic research, please cite the book."
```

```python
res.split(" ")
```

```python
import re
# regular expression
```

```python
# r is used for safer side

re.findall(r'\s',res)
```

```python
# re.findall(r'\s',res)
# it accepts 2 arguments: pattern and location


# to get group of space
re.findall(r'\s+',res) 
# + is used to group the patterns


```

```python
# sub :-- substitute :it is used to replace the patterns
#  it accepts 3 arguments 
# re.sub(r"\s","",res)

```

```python
# if the same of the pattern is 2 or more group it 
# 2 is count, it is more than 2 
#\s{1,2 } it is count from1 to 2
re.sub(r"\s{2,}","",res)

```

```python
# it is used for removing special characters
re.sub(r'[^a-zA-Z0-9]',' ',res)
```
