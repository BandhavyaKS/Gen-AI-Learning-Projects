## **Paragraph based chunking**

```python
with open("1data.txt", "r") as file:
    data = file.read()

paragraphs = data.split("\n\n")

for i, paragraph in enumerate(paragraphs, start=1):
    print(f"Chunk {i}:")
    print(paragraph)
    print("-" * 40)
```

```python
with open("1data.txt", "r") as file:
    data = file.read()

chunks = data.split(". ")

for i in range(len(chunks)):
    print(f"Chunk {i+1}")
    print(chunks[i])
    print("-" * 40)
```
